#### **Q3: How do you distinguish between `inputInt()` and `inputFloat()`?**

**Answer:** 
```
inputInt()   -> Expect integer value as input
inputFloat() -> Expect float value as input
```

#### **Q4: Using PyInputPlus, how do you ensure that the user enters a whole number between 0 and 99?**

**Answer:** Using `min` and `max` argument in `inputInt()` function.
```python
import pyinputplus as pyip
num = pyip.inputInt(min=0, max=99)
```

### **Q5: What is transferred to the keyword arguments `allowRegexes` and `blockRegexes`?**

**Answer:** 
```
allowRegexes : If the regex set in allowRegexes, matches any part of the input, the whole string will be valid.
blockRegexes : If the regex set in blockRegexes matches any part of the input, the whole string will be invalid.
```
