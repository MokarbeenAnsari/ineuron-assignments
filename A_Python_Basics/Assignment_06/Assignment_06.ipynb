#### **Q1: What are escape characters, and how do you use them?**

**Answer:** To insert characters that are illegal in a string or to use their another meaning of that character, we use an escape character.

An escape character is a backslash \ followed by the character you want to insert.

````python
# Exampl 1: We can not print single/double quotes in string when we are surrounding string with single/double quotes
print('Mokarbeen's Laptop') # This will throw error
````

````python
# Escape charcater \ will help here
print('Mokarbeen\'s Laptop')
````

````python
# Example 2: When we use n in between string this will print n
print('Mokarbeen n Ansari')
````

````python
# But when we combine this with backslash \n now n have special meaning i.e. new line character
print('Mokarbeen \n Ansari')
````

#### **Q2: What do the escape characters n and t stand for?**

#### **Q5: How do you write a string of newlines if you don't want to use the n character?**

**Answer:** We can use multiple print statement or **`os.linesep`** keyword.

````python
# Example
import os
print("Mokarbeen" + os.linesep + "Ansari")
````

#### **Q6: What are the values of the given expressions?**
```python
'Hello, world!'[1]
'Hello, world!'[0:5]
'Hello, world!'[:5]
'Hello, world!'[3:]
```

**Answer:**
```python
'Hello, world!'[1]    # 'e'
'Hello, world!'[0:5]  # 'Hello'
'Hello, world!'[:5]   # 'Hello'
'Hello, world!'[3:]   # 'lo, world!'
```

#### **Q7: What are the values of the following expressions?**

#### **Q8: What are the values of the following expressions?**
```python
'Remember, remember, the fifth of July.'.split()
'-'.join('There can only one.'.split())
```

**Answer**:
```python
'Remember, remember, the fifth of July.'.split() # ['Remember,', 'remember,', 'the', 'fifth', 'of', 'July.']
'-'.join('There can only one.'.split())          # 'There-can-only-one.'
```
