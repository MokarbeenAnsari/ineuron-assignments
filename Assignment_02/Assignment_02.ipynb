### **10. How can you tell the difference between break and continue?**

**Answer:** Generally these two keyword used in loop block.

If `break `keyword encountered execution of rest of the block skipped and terminate the loop. Basically `break` is used to terminate the loop at specific condition.

if `continue` keyword encountered execution of rest of the block skipped and execution start from the beginning of block. Basically `continue` used to skip the rest of the block execution at specific condition. This will not terminate the loop.

### **12. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop**.

**Answer:**

````python
# Using for loop
for i in range(1, 11):
  print(i, end=' ')
print()

# Using while loop
i = 1
while i < 11:
  print(i, end=' ')
  i = i + 1
print()
````
